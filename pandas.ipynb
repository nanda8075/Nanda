```python
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'John'],
    'Age': [25, np.nan, 30]
}

df = pd.DataFrame(data)

print(df.isnull())
print(df.fillna(0))
print(df.dropna())
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'John'],
    'Age': [25, 30, 35],
    'Gender': ['F', 'M', 'M']
}

df = pd.DataFrame(data)

print(df.groupby('Gender')['Age'].mean())
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'key': [1, 2],
    'Name': ['Alice', 'Bob']
})

df2 = pd.DataFrame({
    'key': [1, 2],
    'Age': [25, 30]
})

print(pd.merge(df1, df2, on='key'))
```
