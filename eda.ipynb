```python
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris()

iris = pd.DataFrame(
    iris_data.data,
    columns=iris_data.feature_names
)

# Add species column
iris['species'] = pd.Categorical.from_codes(
    iris_data.target,
    iris_data.target_names
)

print(iris['species'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

sns.histplot(iris['sepal length (cm)'], kde=True, ax=axes[0, 0], color='green')
axes[0, 0].set_title('Sepal Length Distribution')

sns.histplot(iris['sepal width (cm)'], kde=True, ax=axes[0, 1], color='skyblue')
axes[0, 1].set_title('Sepal Width Distribution')

sns.histplot(iris['petal length (cm)'], kde=True, ax=axes[1, 0], color='orange')
axes[1, 0].set_title('Petal Length Distribution')

sns.histplot(iris['petal width (cm)'], kde=True, ax=axes[1, 1], color='red')
axes[1, 1].set_title('Petal Width Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris.drop('species', axis=1))
plt.title('Boxplot of All Numerical Features')
plt.ylabel('Centimeters (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
species = df['Species'].unique()
colors = ['red', 'green', 'blue']
# Scatter plot for each species
for i in range(len(species)):
 x = df[df['Species'] == species[i]]
plt.scatter(x['SepalLengthCm'], x['SepalWidthCm'], c=colors[i], label=species[i])
# Add labels and legend
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('Sepal Length vs Width by Species')
plt.legend()
# Show plot
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Sepal Length
sns.boxplot(x='species', y='sepal length (cm)', data=iris, ax=axes[0,0], palette='Set2')
axes[0,0].set_title('Sepal Length by Species')

# Sepal Width
sns.boxplot(x='species', y='sepal width (cm)', data=iris, ax=axes[0,1], palette='Set2')
axes[0,1].set_title('Sepal Width by Species')

# Petal Length
sns.boxplot(x='species', y='petal length (cm)', data=iris, ax=axes[1,0], palette='Set2')
axes[1,0].set_title('Petal Length by Species')

# Petal Width
sns.boxplot(x='species', y='petal width (cm)', data=iris, ax=axes[1,1], palette='Set2')
axes[1,1].set_title('Petal Width by Species')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.violinplot(
    x='species',
    y='petal length (cm)',
    data=iris,
    inner='quartile',
    palette='pastel'
)

plt.title('Petal Length Distribution by Species')
plt.show()
```

```python
from pandas.plotting import andrews_curves
plt.figure(figsize=(10, 6))
andrews_curves(iris, 'species', colormap='viridis')
plt.title('Andrews Curves for Iris Dataset')
plt.show()
```

```python
from pandas.plotting import parallel_coordinates
plt.figure(figsize=(10, 6))
parallel_coordinates(iris, 'species', colormap='viridis')
plt.title('Parallel Coordinates Plot for Iris Dataset')
plt.show()
```
