```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title("my first plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y1=[1,4,9]
y2=[1,2,3]
plt.plot(x,y1,label="squares")
plt.plot(x,y2,label="linear")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
categories=['A','B','C']
values=[5,7,3]
plt.bar(categories,values)
plt.show()
```

```python
import matplotlib.pyplot as plt
data=[1,2,3,3,3,4,4,5]
plt.hist(data,bins=5)
plt.show()
```
